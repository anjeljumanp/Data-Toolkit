2. How does broadcasting work in NumPy?
  - Broadcasting in NumPy allows us to perform arithmetic operations on arrays of different shapes without reshaping them.
  - It automatically adjusts the smaller array to match the larger array's shape by replicating its values along the necessary dimensions.
  - This makes element-wise operations more efficient by reducing memory usage and eliminating the need for loops.
  - Broadcasting applies specific rules to find whether two arrays can be aligned for operations or not that are:
    - Check Dimensions: Ensure the arrays have the same number of dimensions or expandable dimensions.
    - Dimension Padding: If arrays have different numbers of dimensions the smaller array is left-padded with ones.
    - Shape Compatibility: Two dimensions are compatible if they are equal or one of them is 1.

6. What are the differences between NumPy arrays and Python lists?
 - **Data Type Homogeneity**:
    - Python List: Can store elements of different data types (heterogeneous).
    - NumPy Array: Must store elements of the same data type (homogeneous). This allows for more efficient memory storage and faster operations.
 - **Performance**:
   - Python List: Generally slower for numerical operations, especially on large datasets, as operations are performed element by element in Python.
   - NumPy Array: Significantly faster for numerical and mathematical operations due to optimized C implementations and vectorized operations.
 - **Memory Efficiency**:
   - Python List: Less memory efficient, as each element is stored as a separate Python object, requiring additional overhead for type information and references.
   - NumPy Array: More memory efficient, especially for large datasets, as elements of the same type are stored in contiguous memory blocks.
 - **Functionality:**
   - Python List: Offers general-purpose list manipulation, including appending, inserting, and removing elements.
   - NumPy Array: Provides extensive mathematical functions and operations optimized for numerical computations, including linear algebra, Fourier transforms, and random number generation. It also supports broadcasting and shape manipulation.
 - **Fixed vs. Dynamic Size**:
   - Python List: Dynamic size; can grow or shrink as elements are added or removed.
   - NumPy Array: Fixed size; the size must be specified at creation and cannot be changed without creating a new array.

8. What does the term “vectorized operation” mean in NumPy?
  -In NumPy, a vectorized operation refers to performing operations on entire arrays at once, rather than iterating through individual elements using explicit Python loops. This approach leverages NumPy's underlying C and Fortran implementations, resulting in significantly faster and more efficient computations, especially for large datasets.
  - Instead of writing a for loop to apply an operation to each element of an array, you can directly apply the operation to the NumPy array itself. NumPy then handles the element-wise application internally using optimized, pre-compiled code.

15. How does NumPy handle multidimensional arrays?
   - NumPy handles multidimensional arrays using a special structure called the ndarray, which stores data in a continuous block of memory.
   - Each array has a shape, which tells NumPy how many elements exist along each dimension.
   - For example, a shape of (3, 4) means 3 rows and 4 columns, while a shape of (2, 3, 4) represents a 3-dimensional array.
   - NumPy accesses elements using multi-indexing, meaning you can specify one index per dimension, such as arr[i, j] for 2D or arr[x, y, z] for 3D. It also allows slicing across dimensions, letting you extract sub-arrays easily.
   - A key feature is broadcasting, which lets NumPy perform operations between arrays of different shapes by automatically expanding the smaller array without copying data.
   - NumPy uses strides, which define how many bytes to skip to move to the next element in each dimension. This enables fast slicing and view creation.

17. Explain the difference between apply() and map() in Pandas.
  - In Pandas, map() and apply() are used for applying functions, but they differ in where and how they work.
  - The map() function works only on a Series and applies a function element-wise to each value.
  - It is mainly used for simple value transformations or for replacing values using a dictionary or mapping.
  - On the other hand, apply() works on both Series and DataFrame. When used on a Series, it behaves similarly to map() by applying a function to each element.
  - When used on a DataFrame, it can apply a function row-wise or column-wise, making it more flexible for complex operations.
  - While map() is faster for simple tasks, apply() is more powerful because it can handle more advanced row/column computations.
  - map() is ideal for straightforward element-level changes, whereas apply() is suited for broader and more complex transformations.

21. Why is NumPy’s array slicing faster than Python’s list slicing?
  - NumPy’s array slicing is faster than Python’s list slicing because NumPy stores data in a single continuous block of memory, while Python lists store elements as separate objects scattered across memory.
  - When you slice a NumPy array, it does not create a new copy of the data. Instead, it creates a view that points to the same underlying memory, so slicing happens almost instantly.
  - slicing a Python list creates a completely new list with copied elements, which takes more time.
  - NumPy is also built on optimized C code, so operations like slicing and accessing elements run at low-level speeds.
  - Combined with compact memory storage and fewer overheads, NumPy can perform slicing much faster than regular Python lists.

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Marks1": [85, 90, 78, 92, 88],
    "Marks2": [80, 95, 70, 85, 90],
    "Marks3": [75, 88, 82, 91, 87]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")

plt.title("Correlation Matrix Heatmap")
plt.show()
```

11. Perform matrix multiplication using NumPy.

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

C = np.dot(A, B)

C2 = A @ B

print("Result using np.dot():\n", C)
print("Result using @ operator:\n", C2)
```
